2ï¸âƒ£ Find the Largest Element in a List

```python
def find_max(lst):
    max_val = lst[0]
    for num in lst: # Iterate through each number in the list
        if num > max_val: # Check if the current number is greater than max_val
            max_val = num #Update max_val if current number is greater
    return max_val
```

```python
def second_largest(lst):
    unique_lst = list(set(lst))
    unique_lst.sort()
    return unique_lst[-2]
```

5ï¸âƒ£ Count Frequency of Elements

```python
def elemet_frequency(lst):
    freq = {} # Initialize an empty dictionary to store frequencies
    for item in lst:
        if item in freq:
            freq[item] += 1 # Increment count if item already in dictionary
        else:
            freq[item] = 1 # Initialize count to 1 if item not in dictionary
    return freq
```

6ï¸âƒ£ Check if a String is Palindrome

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "") # Normalize the string
    return s == s[::-1]
```

7ï¸âƒ£ Count Vowels in a String

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
```

8ï¸âƒ£ Reverse Words in a String

```python
def reverse_words(sentence):
    words = sentence.split()
    reversed_words = words[::-1]
    return ' '.join(reversed_words) 
print(reverse_words("Hello World from Python"))  # Output: "Python from World Hello"
```

9ï¸âƒ£ Find the Most Frequent Character

```python
def most_frequent_character(lst):
    freq = {} # Initialize an empty dictionary to store frequencies
    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    most_frequent = max(freq, key=freq.get)
    return most_frequent

print(most_frequent_character(['a', 'b', 'c', 'a', 'b', 'a']))  # Output: 'a'
```

```python
def are_anagrams(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    return sorted(str1) == sorted(str2) 
print(are_anagrams("listen", "silent"))  # Output: True
print(are_anagrams("hello", "world"))    # Output: False
```

2. Explain the difference between shallow copy and deep copy in Python.

- A shallow copy creates a new object, but it does not recursively copy nested objects. Instead, it copies references to the same nested objects from the original.

- A deep copy creates a completely independent clone of the entire object, including all nested objects, recursively.

```python
import copy

original_list = [1, [2, 3], 4]
shallow_copied_list = copy.copy(original_list)

shallow_copied_list[1][0] = 99  # Modifying a nested object in the copy
print(original_list)            # Output: [1, [99, 3], 4] - Original is affected
```

```python
import copy

original_list = [1, [2, 3], 4]
deep_copied_list = copy.deepcopy(original_list)

deep_copied_list[1][0] = 99  # Modifying a nested object in the copy
print(original_list)         # Output: [1, [2, 3], 4] - Original remains unchanged
```

== (Equality Operator):

- This is the operator typically used when you want to know if two variables hold the same content.

is (Identity Operator):
- Compares the identity of two objects, meaning it checks if two variables refer to the exact same object in memory.
- Returns True if both variables point to the same memory address, and False otherwise.
- This is used when you need to confirm if two references are literally pointing to the same instance of an object.
